```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Pre-Processing

```python
#Read Dataset
#data = pd.read_excel('data_user_edit.xlsx')
data = pd.read_excel('data_user.xlsx')
data.head(726)
```

```python
data.isnull().sum().sort_values(ascending=False)
data_no_missing = data.dropna(axis=0)
print(data_no_missing.head(5))
```

# Data Integration

```python
data.info()
```

# Konversi tipe data nominal ke tipe numerik

```python
data.drop('User_Id',axis = 1,inplace=True)

konversi_LSI = {"LSI": {1: "Audio", 2: "Kinestetik", 3: "Visual"}}
konversi_Status = {"Status": {1: "Lulus", 8: "Tidak Lulus"}}
data = data.replace(konversi_LSI)
data = data.replace(konversi_Status)
data.head(726)
```

# Data Reduction

```python
data.drop('NIM',axis = 1,inplace=True)
data.drop('Nama',axis = 1,inplace=True)
data.head(726)
```

# EDA - Grafik Data Kelulusan

```python
colors = {'tab:blue', 'tab:cyan'}
label = {'Lulus', 'Tidak Lulus'}

fig, ax = plt.subplots(figsize=(20, 7))
ax.pie(data.Status.value_counts(),labels=label, colors = colors, autopct= '%.0f%%')
ax.set_title('Data Kelas Pada Dataset Kelulusan Matakuliah', fontsize = 20)
plt
```
